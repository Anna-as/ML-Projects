# Deciision Tree

```python
import pandas as pd
```

```python
data=pd.read_csv(r'regressive.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
data=pd.get_dummies(data,columns=['3D_available','Genre'])
```

```python
data.info()
```

```python
data['Time_taken']=data['Time_taken'].fillna(data['Time_taken'].mean())
```

```python
data.info()
```

```python
y=data['Collection']
```

```python
y
```

```python
x=data.drop(['Collection'],axis=1)
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.15)
```

```python
from sklearn import tree
model=tree.DecisionTreeRegressor(max_depth=3)
```

```python
model.fit(xtrain,ytrain)
```

```python
ypredict=model.predict(xtest)
```

```python
ypredict
```

```python
dot=tree.export_graphviz(model,out_file=None)
```
